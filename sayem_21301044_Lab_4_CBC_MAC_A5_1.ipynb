```python
import os
import sys
import binascii

from cryptography.hazmat.primitives import cmac
from cryptography.hazmat.decrepit.ciphers import algorithms

message = "Garlic ice-cream was her favorite"
message_bytes = message.encode()  # Convert string to bytes
key = b'\x05\xf9\x83\x9d\xb7\xb6\xc3\xb8\x9e\xc5\xd9\xd8\x07]\xc6\xb3'
# You code here. HINT: use CMAC, TripleDES
c = cmac.CMAC(algorithms.TripleDES(key))
c.update(message_bytes)
signature = c.finalize()


print(f"Message: {message}")
print(f"CMAC signature: {binascii.b2a_hex(signature).decode()}")
```

```python
#verify the given information here
given_signature = "efd1c31dd97e90a6"

if signature.hex() == given_signature:
    print("Valid")
else:
    print("Invalid")
```

```python
def A51(X, Y, Z, n):

    X = [int(i) for i in X]
    Y = [int(i) for i in Y]
    Z = [int(i) for i in Z]

    keystream = []

    for _ in range(n):

        m = (X[8] & Y[10]) | (X[8] & Z[10]) | (Y[10] & Z[10])

        if X[8] == m:
            new_bit = X[13] ^ X[16] ^ X[17] ^ X[18]
            X = [new_bit] + X[:-1]

        if Y[10] == m:
            new_bit = Y[20] ^ Y[21]
            Y = [new_bit] + Y[:-1]

        if Z[10] == m:
            new_bit = Z[7] ^ Z[20] ^ Z[21] ^ Z[22]
            Z = [new_bit] + Z[:-1]

        key_bit = X[18] ^ Y[21] ^ Z[22]
        keystream.append(key_bit)

    return keystream
```

```python
def text_to_binary(text):
    return ''.join(format(ord(i),'08b') for i in text)
def binary_to_text(b):
    chars=[b[i:i+8] for i in range(0,len(b),8)]
    return ''.join(chr(int(c,2)) for c in chars)
def xor_binary(a,b):
    return ''.join('0' if i==j else '1' for i,j in zip(a,b))
```

```python
X="1110001100101001011"
Y="0011000000010000001101"
Z="10011101101111001001110"

plaintext="Love means never having to say you're sorry"
plain_bin = text_to_binary(plaintext)

keystream = A51(X,Y,Z,len(plain_bin))
keystream = ''.join(str(i) for i in keystream)

ciphertext = xor_binary(plain_bin, keystream)

decrypted_bin = xor_binary(ciphertext, keystream)
decrypted_text = binary_to_text(decrypted_bin)

print("Key stream")
print(keystream)

print("\nCiphertext")
print(ciphertext)

print("\nPlaintext ")
print(decrypted_text)
```
