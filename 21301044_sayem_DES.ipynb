```python

from Crypto.Cipher import DES
from Crypto.Random import get_random_bytes
```

```python
def pad_text(text):
    block_size = 8


    if isinstance(text, str):
        text = text.encode()

    padding_len = block_size - (len(text) % block_size)


    if padding_len == 0:
        padding_len = block_size


    padding = bytes([padding_len] * padding_len)

    padded_text = text + padding

    return padded_text
```

```python
def unpad_text(padded_text):
    padding_len = padded_text[-1]
    return padded_text[:-padding_len]
```

```python
def des_encrypt(plaintext, key):
    if isinstance(plaintext, str):
        plaintext = plaintext.encode()

    if isinstance(key, str):
        key = key.encode()

    cipher = DES.new(key, DES.MODE_ECB)
    ciphertext = cipher.encrypt(plaintext)

    return ciphertext
```

```python
plaintext = b"This is a secret text"

# Generate a random 64-bit (8-byte) key
key = get_random_bytes(8)

print("Original plaintext:", plaintext.decode('utf-8'))
print("Original key :", key)

plaintext = pad_text(plaintext)

ciphertext = des_encrypt(plaintext, key)
print("Encrypted ciphertext", ciphertext)
```

```python
def des_decrypt(ciphertext, key):
    cipher = DES.new(key, DES.MODE_ECB)
    plaintext = cipher.decrypt(ciphertext)
    return plaintext
```
